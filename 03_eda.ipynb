**00 Importing libraries**

```python
# First, we import Pandas ,NumPy ,seaborn and matplotlib, which are the core libraries for data
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset for EDA
insurance_features=pd.read_csv("/content/sample_data/insurance_features.csv")
print(insurance_features.head())
```

**01    Histograms / KDE**

*   'charges'
*   'age'
*   'bmi'





```python
#Q\what does it tell us?
# A\It tells us,Histogram of charges  with kde
sns.histplot(insurance_features["charges"],bins=25 ,kde=True ,color="red")
plt.title("Histogram of charges")
plt.xlabel("medical charges")
plt.ylabel("Frequency")
plt.show()
print("\nthe shape is :\nskewed right")
```

```python
#Q\what does it tell us?
# A\It tells us,Histogram of age with kde
sns.histplot(insurance_features["age"],bins=25 ,kde=True ,color="blue")
plt.title("Histogram of age")
plt.xlabel("age")
plt.ylabel("Frequency")
plt.show()
print("\nthe shape is :\nskewed right")
```

```python
#Q\what does it tell us?
# A\It tells us,Histogram of bmi with kde
sns.histplot(insurance_features["bmi"],bins=25 ,kde=True ,color="green")
plt.title("Histogram of bmi")
plt.xlabel("bmi")
plt.ylabel("Frequency")
plt.show()
print("\nthe shape is :\nsymmetric")
```

**02   Grouped boxplots**

*   'sex_female'
*   'smoker_no'



```python
#Q\what does it tell us?
# A\It tells us, Comparison between Male and Female
sns.boxplot(data=insurance_features,x="sex_female",y="charges",color="Yellow")
plt.title("Boxplot of charges by sex")
plt.xlabel("sex_female")
plt.ylabel("charges")
plt.show()
```

```python
#Q\what does it tell us?
# A\It tells us,Comparison between smokers and non-smokers
sns.boxplot(data=insurance_features,x="smoker_no",y="charges",color="Purple")
plt.title("Boxplot of charges by smoker")
plt.xlabel("smoker")
plt.ylabel("charges")
plt.show()
```

**03   Correlation heatmap**

```python
# Q\what does it tell us?
#A\It tells us, what are the top 10 features most correlated with the target variable

only_features=insurance_features.drop(columns=["charges"]) # the dataset Except the target
plt.figure(figsize=(10, 10))
sns.heatmap(only_features.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation  heatmap")
plt.show()
```

**04   Scatter plot**

```python
# Q\what does it tell us?
# A\It tells us,  the Relationship between Age , Charges and Smoker
sns.scatterplot(data=insurance_features, x='age', y='charges',hue="smoker_no", palette='magma', alpha=1)
# i use color for a third variable
plt.title('Relationship between Age , Charges and Smoker') # the title of scatterplot
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.show()
```

**05    Groupby summary**

*   'sex_famele'
*   'smoker_no'
*   'region'
*   'High risk'




```python
def Groupby_summary(df):
  # Create function for Create subplots for each axes
  fig, axes = plt.subplots(2, 2, figsize=(10, 8)) # Create a 2x2 grid of subplots
  fig.suptitle('Groupby Summary') # the title of fig

# 01 creat barplot of 'sex_female'
# Q\what does it tell us?
#A\It tells us,Average Charges by Gender
  sns.barplot(data=df, x='sex_female', y='charges', palette='Reds',ax=axes[0, 0])
  axes[0, 0].set_title('Average Charges by Gender')
  axes[0, 0].set_xlabel('sex_female')
  axes[0, 0].set_ylabel('Average Charges')

# 02 creat barplot of 'smoker_no'
# Q\what does it tell us?
#A\It tells us,Average Charges by smoker
  sns.barplot(data=df, x='smoker_no', y='charges', palette='Reds',ax=axes[0, 1])
  axes[0, 1].set_title('Average Charges by smoker')
  axes[0, 1].set_xlabel('smoker_no')
  axes[0, 1].set_ylabel('Average Charges')

# 03 creat barplot of 'region'
# Q\what does it tell us?
#A\It tells us,Average Charges by Region
  sns.barplot(data=df, x='region', y='charges', palette='Reds',ax=axes[1, 0])
  axes[1, 0].set_title('Average Charges by Region')
  axes[1, 0].set_xlabel('Region')
  axes[1, 0].set_ylabel('Average Charges')

# 04 creat barplot of 'High Risks'
# Q\what does it tell us?
#A\It tells us,Average Charges by High Risks
  sns.barplot(data=df, x='High Risks', y='charges', palette='Reds',ax=axes[1, 1])
  axes[1, 1].set_title('Average Charges by High Risks')
  axes[1, 1].set_xlabel('High Risks')
  axes[1, 1].set_ylabel('Average Charges')

  plt.tight_layout() # adjusts spacing
  plt.show()

print(Groupby_summary(insurance_features))
```

```python
# Q\which group is highest / lowest?
# A\ we can know by using groupby
print("\nGroupby Summary: Mean Charges by Category:\n",insurance_features.groupby("sex_female")["charges"].mean())
print("\n\nGroupby Summary: Mean Charges by Category:\n",insurance_features.groupby("smoker_no")["charges"].mean())
print("\n\nGroupby Summary: Mean Charges by Category:\n",insurance_features.groupby("region")["charges"].mean())
print("\n\nGroupby Summary: Mean Charges by Category:\n",insurance_features.groupby("High Risks")["charges"].mean())
```
