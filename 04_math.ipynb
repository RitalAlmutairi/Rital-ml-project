**00 Importing libraries**

```python
# First, we import Pandas and NumPy, which are the core libraries for data
import pandas as pd
import numpy as np
```

```python
# Loading the dataset for EDA
insurance_features=pd.read_csv("/content/sample_data/insurance_features.csv")
print(insurance_features.head())
```

**01   • Compute the mean and standard deviation**

```python
# mean
total =insurance_features["charges"].sum() # total charges
count=len(["charges"]) # count charges
mean_target=total/count # mean

print("\nthe mean of the target:\n",mean_target)


# std
differences=insurance_features["charges"]-mean_target # differences
square=differences**2 # squared differences
variance = np.sum(square) / count # Variance
std_target= np.sqrt(variance) # std

print("\nthe mstd of the target:\n",std_target)
```

**02   Standardise**

```python
# z = (X - mean) / std —

"""by hand"""
# mean
total=insurance_features["age"].sum()# total age
count=len(insurance_features["age"]) # count age
mean_age=total/count # mean age
# std
Subtract=insurance_features["age"]-mean_age # Subtract
square=Subtract**2 # squared differences
variance = np.sum(square) / count # Variance
std_age= np.sqrt(variance) # std
z_score = Subtract/std_age# Division

print("\nby hand :\n",z_score)


"""using StandardScaler"""

mean_age=insurance_features["age"].mean() # mean
std_age=insurance_features["age"].std()# std
Z_score=(insurance_features["age"]-mean_age)/std_age # Z_score
print("\n\nusing StandardScaler :\n",Z_score)
```

**03   Compute cosine similarity**

```python
# import
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

numric_features = ["age", "bmi" , "children","charges_age_ratio","age_bmi","charges_log1p"] #get  only numric features

high = insurance_features["charges"].idxmax() # get the high indices
low = insurance_features["charges"].idxmin() # get the low indices

# StandardScaler
scaler = StandardScaler()
insurance_features[numric_features] = scaler.fit_transform(insurance_features[numric_features])
scaled_features = insurance_features[numric_features]

# vectors
vectors_high = scaled_features.loc[high].values
vectors_low = scaled_features.loc[low].values

# Cosine Similarity
csim = np.dot(vectors_high, vectors_low) / (np.linalg.norm(vectors_high) * np.linalg.norm(vectors_low))
print(f"index High: {high},\n index Low: {low}\nCosine Similarity: {csim:.4f} ")
```

**04   Estimate a probability**

```python
# threshold is the neam of ["charges"]
threshold = 13229.380903755422

# Smokers only
smokers = insurance_features[insurance_features['smoker_no'] == 1]

# number of smokers & the threshold
smokers_threshold = smokers[smokers['charges'] > threshold]

# the part over the whole
Empirical_probability = len(smokers_threshold) / len(smokers)

print(f"total Smokers: {len(smokers)}\nsmokers above threshold: {len(smokers_threshold)}\nProbability: {Empirical_probability:.4f}")
```
